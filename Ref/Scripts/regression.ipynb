# REGRESSION

```r
df <- read.csv("/finalData.csv")
```

```r
df_ts <- ts(df, start = c(2012,121.25), frequency = 365.25)
```

Checking for any variance that is present in the data

```r
df_ts[,'temp'] <- df_ts[,'temp'] + 50
```

```r
which.min(df_ts[,'temp'])
df_ts[1552,'temp']
```

```r
BoxCox.lambda(df_ts[,'temp'])
```

From the boxcox results we can say that there is some variance that is present and can be stabilized by uisng lambda value of 2.0.

data wrangle

```r
library(tidyverse)
```

```r
df_reg <- df %>% select(-feelslikemax,
                        -feelslikemin,
                        -feelslike,
                        -name,
                        -conditions,
                        -description,
                        -stations,
                        -sunset,
                        -sunrise,
#                        -datetime,
                        -sunrise,
                        -sunset,
                        -preciptype,
                        -icon,
                        -datetime,
                        -avgTemp)
```

```r
df_reg$windgust <- ifelse(is.na(df_reg$windgust), 0, df_reg$windgust)
df_reg$severerisk <- ifelse(is.na(df_reg$severerisk), 0, df_reg$severerisk)
```

Explore the best model for temp

```r
df_temp <- df_reg %>% select(-tempmax, -tempmin)
```

```r
df_temp_ts <- ts(df_temp, start = c(2012, 121.25), frequency = 365.25)
```

```r
train_temp <- window(df_temp_ts, start = c(2012, 121.25), end = c(2021, 120))
```

```r
test_temp <- window(df_temp_ts, start = c(2021, 121))
```

Forward stepwise

```r
library(leaps)
```

```r
regfit <- regsubsets(temp ~., 
                     data = train_temp, 
                     nvmax = 19,
                     method = "forward")
```
